# Elbow method

Đây là một phương pháp giúp ta chọn ra được số cụm cần thiết để phân cụm, để tránh việc phân quá nhiều cụm dẫn tới độ phức tạp của mô hình tăng và khả năng cao bị overfitting do có quá nhiều cụm.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

X, y = make_blobs(
        centers=3, n_samples=500, n_features=2, shuffle=True, random_state=40)
```

```python
plt.scatter(x=X[:,0], y=X[:,1])
```

```python
model = KMeans(n_clusters=3)
model.fit(X).predict(X)
model.inertia_
```

```python
inertias = []

for i in range(1, 11):
    model = KMeans(n_clusters=i)
    model.fit(X).predict(X)
    inertias.append(model.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(range(1,11))
plt.show()
```

Ta thấy sử dụng phương pháp Elbow cho ta số lượng tâm cụm cần khởi tạo là bằng 3.

## Ví dụ với sklearn

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd
```

```python
df = load_breast_cancer()
```

```python
X, y = df.data, df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
y_pred
```

```python
y_pred_2d = y_pred.reshape(len(y_pred), 1)
y_test_2d = y_test.reshape(len(y_test), 1)
```

```python
df1 = pd.DataFrame(y_pred_2d)
df2 = pd.DataFrame(y_test_2d)
```

```python
total_df = df1.join(df2,lsuffix='x', rsuffix='y')
```

```python
total_df.columns = ['pred', 'real']
```

```python
total_df[total_df['pred'] == total_df['real']].shape[0] / len(total_df)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df = load_breast_cancer()
X = df.data
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

reg_lasso = Lasso(alpha = 1.0)
reg_lasso.fit(X_train, y_train)

# Sai số huấn luyện trên tập train
print(reg_lasso.score(X_train, y_train))

# Hệ số hồi qui và hệ số chặn
print(reg_lasso.coef_)
print(reg_lasso.intercept_)
```

```python
y_pred = reg_lasso.predict(X_test)
mean_squared_error(y_pred, y_test)
```

Hình minh họa trên cho ta thấy sự khác biệt về miền xác định do ràng buộc của thành phần điều chuẩn của Lasso và Ridge. Lasso là hình thoi, Ridge là hình tròn.

Nghiệm của hồi quy Lasso là điểm tiếp xúc giữa đường đồng mức của hàm mất mát hồi quy tuyến tính (hình elip) và miền xác định 
L1 (hình thoi). Cơ chế này giúp Lasso thực hiện lựa chọn biến và làm cho một số hệ số hồi quy bằng 0. Còn Ridge thì ta thấy điểm tiếp xúc chỉ ở đâu đố gần điểm 0.

Nhưng câu hỏi đặt ra là nếu nghiệm xa cực tiểu của hàm mất mát của hồi quy tuyến tính thì làm sao đảm bảo hàm mất mát đạt giá trị nhỏ nhất?

- **Hàm mất mát không bị tối thiểu hóa tuyệt đối**: Trong hồi quy Lasso, hàm mất mát (thường là tổng bình phương sai số - RSS) không được tối thiểu hóa một cách tuyệt đối mà bị tối ưu hóa dưới ràng buộc. Điều này có nghĩa là chúng ta không chỉ tìm kiếm nghiệm tối thiểu hóa hàm mất mát mà còn phải thỏa mãn ràng buộc L1 lên các hệ số.

- **Cân bằng giữa độ chính xác và độ phức tạp**: Khi thêm ràng buộc L1, chúng ta đang tìm kiếm một sự cân bằng giữa độ chính xác của mô hình (thể hiện qua hàm mất mát) và độ phức tạp của mô hình (thể hiện qua số lượng các hệ số khác 0). Ràng buộc này dẫn đến việc chúng ta có thể chọn một nghiệm gần điểm cực tiểu của hàm mất mát nhưng có nhiều hệ số bằng 0, giúp mô hình dễ dàng hơn và tránh overfitting.

- **Ý nghĩa của điểm tiếp xúc**: Điểm tiếp xúc giữa hình elip của hàm mất mát và hình thoi của ràng buộc Lasso không nhất thiết phải là điểm cực tiểu tuyệt đối của hàm mất mát.

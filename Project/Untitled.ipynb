## Regular Expression

```python
import re

regex = '[a]'
matches = re.finditer(regex, 'abc')
for i in matches:
    print(i)
```

```python
re.match(regex, 'a')
```

```python
my_re = re.compile('([0-9]+)[^0-9]+([0-9]+)')
m = my_re.search('Anna is 15 years old and John is 12 years old.')
print(m.group(0))
print(m.group(1))
print(m.group(2))
```

```python
my_re = re.compile(r'\d')
matches = my_re.finditer('Anna is 15 years old and John is 12 years old.')
for match in matches:
    print(match)
```

```python
if re.match(my_re, 'Anna is 15 years old and John is 12 years old.'):
    print('True')
```

```python
worf_quote = "Sir, I protest. I am not a merry man!"
words_in_quote = word_tokenize(worf_quote)
words_in_quote
```

```python
stop_words = set(stopwords.words('english'))
filtered_list = []

for word in words_in_quote:
    if word.casefold() not in stop_words:
        filtered_list.append(word)
```

```python
filtered_list
```

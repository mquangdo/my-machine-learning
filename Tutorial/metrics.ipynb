# Metrics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('creditcard.csv')[:80000]
```

```python
df['Class'].value_counts()
```

```python
X = df.drop(columns=['Time','Amount','Class']).values
y = df['Class'].values
```

```python
X, y
```

```python
f'Shape of X: {X.shape}, y: {y.shape}, Fraud Cases: {y.sum()}'
```

Ta thấy trên tổng 80000 records mà chỉ có 196 là trường hợp lừa đảo.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns

mod = LogisticRegression()
pred = mod.fit(X, y).predict(X)
accuracy_score(y, pred)
sns.heatmap(confusion_matrix(pred, y), annot = True)
plt.xlabel('Pred')
plt.ylabel('True')
```

```python
pred.sum()
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=LogisticRegression(), param_grid={'class_weight':[{0:1,1:v} for v in range(1, 4)]}, cv=4, n_jobs=-1)
grid.fit(X, y)
```

```python
pd.DataFrame(grid.cv_results_)
```

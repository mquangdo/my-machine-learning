Còn với các biến phân loại không có thứ tự thì ta thường sử dụng OneHotEncoder.

Các thuật toán: Logistic Regression, Neural Networks, K-Nearest Neighbors (KNN), Support Vector Machines (SVM), Naive Bayes, Principal Component Analysis (PCA), Clustering (K-Means, etc.) thì ta thường xử lí các biến phân loại bằng OneHotEncoder.


Việc sử dụng OneHotEncoder cho KNN là cần thiết vì cách KNN hoạt động dựa trên việc tính toán khoảng cách giữa các điểm dữ liệu. Dưới đây là lý do chi tiết:

- 1. Tránh tạo ra thứ tự giả định không tồn tại:
Nếu bạn sử dụng LabelEncoder để mã hóa các giá trị danh mục, mỗi giá trị sẽ được gán một số nguyên (ví dụ: "Red" = 0, "Blue" = 1, "Green" = 2). Điều này tạo ra một thứ tự giả định giữa các giá trị (0 < 1 < 2), dù trên thực tế, không có mối quan hệ thứ tự nào giữa các màu sắc này.
Khi tính toán khoảng cách giữa các điểm, KNN sẽ coi rằng "Red" gần với "Blue" hơn là "Green", chỉ vì các giá trị số của chúng. Điều này có thể dẫn đến kết quả sai lệch, vì mô hình đang giả định có một loại mối quan hệ nào đó mà thực sự không tồn tại.


- 2. Khoảng cách Euclidean:
KNN thường sử dụng khoảng cách Euclidean để tìm các hàng xóm gần nhất. Trong trường hợp các đặc trưng danh mục được mã hóa bằng LabelEncoder, khoảng cách giữa các giá trị danh mục có thể không mang ý nghĩa thực tế.
OneHotEncoder giải quyết vấn đề này bằng cách mã hóa từng giá trị danh mục thành một vector nhị phân, trong đó mỗi giá trị được biểu diễn dưới dạng một cột riêng biệt. Ví dụ, nếu có ba màu sắc ["Red", "Blue", "Green"], chúng sẽ được mã hóa thành ba cột như sau:

                                                    - "Red": [1, 0, 0]
                                                    - "Blue": [0, 1, 0]
                                                    - "Green": [0, 0, 1]

Với mã hóa này, khoảng cách giữa các giá trị danh mục không còn bị ảnh hưởng bởi thứ tự giả định, và KNN có thể tính toán khoảng cách chính xác giữa các điểm dữ liệu.


- 3. Giúp mô hình phản ánh đúng bản chất của dữ liệu:
OneHotEncoder giúp mô hình KNN xem xét tất cả các giá trị danh mục là độc lập và không có mối quan hệ thứ tự. Điều này phản ánh đúng bản chất của các biến danh mục trong thực tế, nơi mà mỗi giá trị chỉ là một nhãn hoặc một loại mà không có sự hơn kém giữa chúng.
Nhờ đó, mô hình KNN có thể phân loại và tìm các điểm dữ liệu tương tự một cách chính xác hơn, dựa trên những gì thực sự tồn tại trong dữ liệu.

Sử dụng hold-out trong thực tế để chia bộ dữ liệu thành các bộ huấn luyện và kiểm tra.

```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data')
```

```python
df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids', 'Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue', 'diluted wines','Proline']
```

```python
df_wine
```

```python
from sklearn.model_selection import train_test_split

X, y = df_wine.iloc[:,1:].values, df_wine.iloc[:, 0].values #X là các dữ liệu còn lại, y là label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=0,stratify=y)
```

Tham số stratify trong hàm train_test_split của scikit-learn được sử dụng để đảm bảo rằng tỷ lệ của các lớp (labels) trong tập huấn luyện và tập kiểm tra tương tự như trong tập dữ liệu gốc. Điều này đặc biệt quan trọng trong các bài toán phân loại, nơi mà dữ liệu có thể bị lệch lớp (class imbalance), tức là một số lớp xuất hiện nhiều hơn hẳn so với các lớp khác.
